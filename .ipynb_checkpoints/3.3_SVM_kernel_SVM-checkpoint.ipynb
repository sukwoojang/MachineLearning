## SVM

* SVM (Support Vector Machine) : 강력하고 널리 사용되는 학습 알고리즘, 퍼셉트론의 확장으로 생각할 수 있으며, 앞서 퍼셉트론 알고리즘을 사용하여 분류 오차를 최소화했다면 SVM의 최적화 대상은 마진을 최대화 하는 것이다.
* 마진 : 마진은 클래스를 구분하는 초평면과 이 초평면에 가장 가까운 훈련 샘플 사이의 거리로 정의. 큰 마진의 결정 경계는 일반화 오차가 낮아지는 경향이 있고 작은 마진의 모델은 과대적합되기가 쉽다.

### 슬랙 변수를 사용하여 비선형 분류 문제 다루기

* 슬랙변수 : 선형적으로 구분되지 않는 데이터에서 선형 제약 조건을 완화할 필요가 있기 때문에 도입. 이를 통해 적절히 비용을 손해 보면서 분류 오차가 있는 상황에서 최적화 알고리즘이 수렴한다.
* 소프트 마진 분류 : 슬랙 변수를 이용한 비선형 분류

![image.png](./image/slack1.jpg)
![image.png](./image/slack2.jpg)

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler

iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.3, random_state = 1, stratify=y)

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, test_idx=None, resolution = 0.02):
    
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    x1_min, x1_max = X[:, 0].min() - 1, X[:,0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:,1].max() + 1
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                          np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha = 0.3, cmap = cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0],
                   y=X[y==cl, 1],
                   alpha =0.8,
                   c = colors[idx],
                   marker = markers[idx],
                   label = cl,
                   edgecolor = 'black')
    if test_idx:
        X_test, y_test = X[test_idx, :], y[test_idx]
        
        plt.scatter(X_test[:,0], X_test[:, 1],
                   c = '', edgecolor = 'black', alpha = 1.0,
                   linewidth = 1, marker = 'o',
                   s = 100, label = 'test set')
```

```python
X_combined_std = np.vstack((X_train_std,X_test_std))
y_combined = np.hstack((y_train, y_test))
```

```python
# SVM 모델 훈련
from sklearn.svm import SVC

svm = SVC(kernel = 'linear', C = 1.0, random_state = 1)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx = range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```

### 커널 SVM을 사용하여 비선형 문제 풀기

* 커널 방법 : 매핑 함수 $ \phi $를 사용하여 원본 특성의 비선형 조합을 선형적으로 구분되는 고차원 공간에 투영하는 것
* 매핑 함수 제약 : 이러한 매핑 방식은 새로운 특성을 만드는 계산 비용이 매우 비싸다. 특히나 고차원일 경우
* 커널 함수 : 두 포인트 ${x_i^T}{x_j},\phi({x_i^T})\phi({x_j})$사이 점곱을 계산하는 데 드는 높은 비용을 절감하기 위한 함수, 샘플 간의 유사도 함수로 해석할 수 있다.지수 함수로 얻게 되는 유사도 점수는 1(매우 비슷한 샘플)과 0(매우 다른 샘플) 사이의 범위를 가지게 된다.
$$ \kappa({x_i^T}{x_j}) = \phi({x_i^T})\phi({x_j})$$
* 방사 기저 함수(Radial Basis Function, RBF) : 가장 널리 사용되는 커널 중 하나로 가우시안 커널이라고도 한다. 
$$ \kappa({x_i^T}{x_j}) = exp(-\gamma\Vert{x_t^T-x_j}\Vert^2) $$
$$ \gamma = \frac{1}{2\sigma^2}$$

![텍스트](https://tensorflowkorea.files.wordpress.com/2017/06/2-37.png?w=768)

![텍스트](https://tensorflowkorea.files.wordpress.com/2017/06/2-38.png?w=768)

```python
# 비선형 데이터셋

np.random.seed(1)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)

plt.scatter(X_xor[y_xor == 1, 0], X_xor[y_xor == 1, 1], c='b', marker ='x', label = '1')
plt.scatter(X_xor[y_xor == -1, 0], X_xor[y_xor == -1, 1], c='r', marker ='s', label = '-1')
plt.xlim([-3,3])
plt.ylim([-3,3])
plt.legend(loc = 'best')
plt.tight_layout()
plt.show()
```

```python
# kernel = 'rbf'
svm = SVC(kernel = 'rbf',gamma = 0.10,  C = 10.0, random_state = 1)
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```

* 매개변수$\gamma$의 역할 : 가우시안 구의 크기를 제한하는 매개변수로 값을 크게 하면 서포트 벡터의 영향이나 범위가 줄어든다. 즉, 결정 경계는 더욱 샘플에 가까워지고 구불구불해진다
<br>(과대적합 얘기를 하는거 같음..)

```python
# gamma 값 비교, iris data set

# gamma = 0.2
svm = SVC(kernel = 'rbf', C = 1.0,gamma = 0.2, random_state = 1)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx = range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```

* 비교적 $\gamma$값을 작게 했기 때문에 RBF 커널 SVM모델이 만든 결정 경계는 부드럽다

```python
# gamma = 100

svm = SVC(kernel = 'rbf', C = 1.0,gamma = 100.0, random_state = 1)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx = range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc = 'upper left')
plt.tight_layout()
plt.show()
```
